```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
#check the data information is there any null value or missing values.
data.info()
```

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
data.hist(figsize = (20,15));
```

```python
data.corr()
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(data.corr(),annot = True ,cmap = "YlGnBu")
```

```python
# let do some preprocessing.
# as we have seen in histogram that following variable are rightly skewed
# will perform on them
# Avg. Area Number of Rooms
data['Avg. Area Number of Rooms']=np.log(data['Avg. Area Number of Rooms']+1)
data['Avg. Area Number of Bedrooms']=np.log(data['Avg. Area Number of Bedrooms']+1)
data['Area Population']=np.log(data['Area Population']+1)
data['Price']=np.log(data['Price']+1)
```

```python
data.hist(figsize=(20,15));
```

```python
#one hot encoding
data.Address.value_counts()
```

```python
data['bedroom_ratio']=data['Avg. Area Number of Bedrooms']/data['Avg. Area Number of Rooms']
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(data.corr(),annot = True ,cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = data.drop(['Price'],axis=1)
Y = data['Price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
linear_model=LinearRegression()
linear_model.fit(X_train,Y_train)
```

```python
linear_model.score(X_test,Y_test)
```
