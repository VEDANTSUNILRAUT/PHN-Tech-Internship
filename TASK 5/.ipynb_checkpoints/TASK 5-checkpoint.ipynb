```python
# 1. Write a Python code to Create a figure with 2x2 subplots.

import matplotlib.pyplot as plt
import numpy as np

#2. In the first subplot (top-left), plot a bar chart using the following data:
#Categories: ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
#Values: [23, 56, 41, 62, 19]

Categories=(['Cat-A', 'Cat-B', 'Cat-C', 'Cat-D','Cat-E'])
Values=([23, 56, 41, 62, 19])
fig, ax=plt.subplots(2,2)
plt.subplot(2,2,1)
plt.title('SUBPLOT:1')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.bar(Categories,Values)

#3.In the second subplot (top-right), plot a histogram using the following data:
#Data: [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

histo = np.random.normal( [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10])
plt.subplot(2,2,2)
plt.title('SUBPLOT:2')
plt.xlabel('DATA')
plt.ylabel('DATA')
plt.hist(histo)

# 4.In the third subplot (bottom-left), plot a pie chart using the following data:
#Labels: ['Apple', 'Banana', 'Orange', 'Mango']
#Sizes: [30, 25, 15, 30]

Labels=['Apple', 'Banana', 'Orange', 'Mango']
Sizes=[30, 25, 15, 30]
v =plt.subplot(2,2,3)
plt.title('SUBPLOT:3')
plt.xlabel('Pie-chart of Fruit')
plt.pie(Sizes, labels=Labels,autopct='%2.1f%%')

# 4.In the fourth subplot (bottom-right), plot a scatter plot using the following data:
# X values: [1, 2, 3, 4, 5]
# Y values: [2, 5, 3, 6, 4]

X_values=[1, 2, 3, 4, 5]
Y_values=[2, 5, 3, 6, 4]
v =plt.subplot(2,2,4)
plt.title('SUBPLOT:4')
plt.xlabel('X_values')
plt.ylabel('Y_values')
plt.scatter(X_values,Y_values)
fig.tight_layout()
plt.show()
```
